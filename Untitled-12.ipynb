#### Create a new column of total_active_minutes by using the columns of VeryActiveMinutes, FairlyActiveMinutes,  LightlyActiveMinutes, SedentaryMinutes

```python
dailyActivity_merged_df['Total_active_minutes'] = dailyActivity_merged_df.VeryActiveMinutes + dailyActivity_merged_df.FairlyActiveMinutes + dailyActivity_merged_df.LightlyActiveMinutes +  dailyActivity_merged_df.SedentaryMinutes
```

```python
dailyActivity_merged_df
```

### Grouping  Total steps, Total Distance and Calories by based on the Week.

```python
dailyActivity_merged_df ['ActivityDate'] = pd.to_datetime(dailyActivity_merged_df.ActivityDate)
```

```python
dailyActivity_merged_df ['ActivityDate']
```

```python
dailyActivity_merged_df['year'] = pd.DatetimeIndex(dailyActivity_merged_df.ActivityDate).year
dailyActivity_merged_df['month'] = pd.DatetimeIndex(dailyActivity_merged_df.ActivityDate).month
dailyActivity_merged_df['day'] = pd.DatetimeIndex(dailyActivity_merged_df.ActivityDate).day
dailyActivity_merged_df['weekday'] = pd.DatetimeIndex(dailyActivity_merged_df.ActivityDate).weekday
```

```python
dailyActivity_merged_df.groupby('weekday')[['TotalSteps', 'TotalDistance', 'Calories']]
```

```python
weekly_groups = dailyActivity_merged_df.groupby ('weekday')
```

```python
weekly_groups [['TotalSteps', 'TotalDistance', 'Calories']].mean()
```

### Write the data to a new work sheet

```python
result_df = dailyActivity_merged_df [['Id', 'ActivityDate', 'TotalSteps', 'TotalDistance', 'Calories', 'Total_active_minutes']]
```

```python
result_df.to_csv('results.csv', index=None)
```
